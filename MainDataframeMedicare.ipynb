```python
# Import Dependencies
import pandas as pd
```

```python
# Read file into DataFrame
medi_df = pd.read_csv('Resources/Medicare_Part_D_Spending_by_Drug_2021.csv')
#medi_df.head(20)
medi_df
```

```python
medi_df.info()
```

```python
#Set an index

medi_index = medi_df.set_index('Gnrc_Name')
medi_index
#Filter for insulin drug (we need only insulin)
medi_filtered_df = medi_index.loc[['Insulin Glulisine','Insulin Lispro', 'Insulin Aspart','Insulin Regular, Human']]

medi_filtered_df
```

```python
#change the spending columns to more user friendly numbers for plotting  
# Define a formatting function to format float values to two decimal places
def format_float(value):
    return '{:.2f}'.format(value)

# Specify the columns to format
columns_to_format = ['Tot_Spndng_2019', 'Tot_Dsg_Unts_2019', 'Tot_Spndng_2020', 'Tot_Dsg_Unts_2020', 'Tot_Spndng_2021', 'Tot_Dsg_Unts_2021']

# Apply the formatting function to the selected columns
medi_filtered_df[columns_to_format] = medi_filtered_df[columns_to_format].applymap(format_float)

medi_filtered_df
```

```python
#getting the column names for the medi_flitered_df to make the plotting easier 
medi_filtered_df.columns
```

```python
#Rename [Tot_Spndng_2019', 'Tot_Dsg_Unts_2019', 'Tot_Spndng_2020', 'Tot_Dsg_Unts_2020', 'Tot_Spndng_2021', 'Tot_Dsg_Unts_2021']

column_mapping = {
    'Tot_Spndng_2019': 'Total Spending 2019',
    'Tot_Dsg_Unts_2019': 'Total Dsg Units 2019',
    'Tot_Clms_2019 ': 'Total Claims 2019',
    'Tot_Spndng_2020': 'Total Spending 2020',
    'Tot_Dsg_Unts_2020': 'Total Dsg Units 2020',
    'Tot_Clms_2020': 'Total Claims 2020',
    'Tot_Spndng_2021': 'Total Spending 2021',
    'Tot_Dsg_Unts_2021': 'Total Dsg Units 2021',
    'Tot_Clms_2021': 'Total Claims 2020'
}

medi_filtered_df.rename(columns=column_mapping, inplace=True)

# columns_to_keep = ['Total Spending 2019', 'Total Dsg Units 2019', 'Total Spending 2020',
#                   'Total Dsg Units 2020', 'Total Claims 2020', 'Total Spending 2021', 'Total Dsg Units 2021', 'Total Claims 2020', 'Tot_Mftr', 'Mftr_Name']

# Select and keep only the specified columns
# medi_filtered_df = medi_filtered_df[columns_to_keep]
medi_filtered_df.head()
```

```python
#Converting spending columns to user friendly option

columns_to_format = ['Total Spending 2019', 'Total Dsg Units 2019',
                     'Total Spending 2020', 'Total Dsg Units 2020',
                     'Total Spending 2021', 'Total Dsg Units 2021']

medi_filtered_df[columns_to_format] = medi_filtered_df[columns_to_format].astype(float)

# Apply the formatting to the specified columns
medi_filtered_df[columns_to_format] = medi_filtered_df[columns_to_format].applymap(lambda x: '{:.2f}'.format(x))

medi_filtered_df.head()
```

```python
#Calculating and plotting the mean for spending for years 2019-2021



import numpy as np

# Replace 'nan' with NaN (missing values)
medi_filtered_df = medi_filtered_df.replace('nan', np.nan)

# Specify the columns to convert to numeric
numeric_columns = ['Total Spending 2019',
                   'Total Spending 2020', 
                   'Total Spending 2021']

# Attempt to convert the columns to numeric
medi_filtered_df[numeric_columns] = medi_filtered_df[numeric_columns].apply(pd.to_numeric, errors='coerce')

#import matplotlib
import matplotlib.pyplot as plt

columns_to_include = ['Total Spending 2019', 
                   'Total Spending 2020', 
                   'Total Spending 2021',]

# list with claims included ['Total Spending 2019', 'Total Dsg Units 2019', 'Total Claims 2019', \
#                    'Total Spending 2020', 'Total Dsg Units 2020', 'Total Claims 2020',\
#                    'Total Spending 2021', 'Total Dsg Units 2021', 'Total Claims 2020']

# Create DF for only the selected columns
medi_filtered_2019_2021 = medi_filtered_df[columns_to_include]

# Calculating the mean for each column
medi_mean_values = medi_filtered_2019_2021.mean()

#Create bar plot for the mean values

plt.figure(figsize=(12, 6))
medi_mean_values.plot(kind='bar')
plt.title('Mean Spending Values (in dollars) for Insulin - Years 2019-2021')
plt.xlabel('Columns')
plt.ylabel('Mean Value')
plt.ticklabel_format(style='plain', axis='y')
plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig('Resources/mean_values_spending_plot.png')
plt.show()
```

```python
#Calculating and plotting the mean for dosages for  years 2019-2021

import numpy as np

# Replace 'nan' with NaN (missing values)
medi_filtered_df = medi_filtered_df.replace('nan', np.nan)

# Specify the columns to convert to numeric
numeric_columns_dsg = ['Total Dsg Units 2019',
                   'Total Dsg Units 2020',
                   'Total Dsg Units 2021']

# Attempt to convert the columns to numeric
medi_filtered_df[numeric_columns_dsg] = medi_filtered_df[numeric_columns_dsg].apply(pd.to_numeric, errors='coerce')

#import matplotlib
import matplotlib.pyplot as plt

columns_to_include_dsg = ['Total Dsg Units 2019',
                   'Total Dsg Units 2020',
                   'Total Dsg Units 2021']

# list with claims included ['Total Spending 2019', 'Total Dsg Units 2019', 'Total Claims 2019', \
#                    'Total Spending 2020', 'Total Dsg Units 2020', 'Total Claims 2020',\
#                    'Total Spending 2021', 'Total Dsg Units 2021', 'Total Claims 2020']

# Create DF for only the selected columns
medi_filtered_2019_2021 = medi_filtered_df[columns_to_include_dsg]

# Calculating the mean for each column
mean_values = medi_filtered_2019_2021.mean()

#Create bar plot for the mean values

plt.figure(figsize=(12, 6))
mean_values.plot(kind='bar', color = 'g')
plt.title('Mean dosage units for Insulin from Medicare - 2019-2021')
plt.xlabel('Columns')
plt.ylabel('Mean Value')
plt.ticklabel_format(style='plain', axis='y')
plt.xticks(rotation=90)
plt.tight_layout()

plt.savefig('Resources/mean_values_dosages_plot.png')
plt.show()
```

```python
#calculating the Median for 2019-2020

# Define the columns you want to include
columns_to_include_median = ['Total Dsg Units 2019',
                   'Total Dsg Units 2020',
                   'Total Dsg Units 2021','Total Spending 2019',
                   'Total Spending 2020', 
                   'Total Spending 2021']

median_columns_spnd = ['Total Spending 2019',
                   'Total Spending 2020', 
                   'Total Spending 2021']

median_columns_dsg = ['Total Dsg Units 2019',
                   'Total Dsg Units 2020',
                   'Total Dsg Units 2021']

# Create DF for spend and one for dsg 
median_filtered_spnd = medi_filtered_df[median_columns_spnd]

median_filtered_dsg = medi_filtered_df[median_columns_dsg]

# Calculate the median for each column
median_values_spnd = median_filtered_spnd.median()

median_filtered_dsg = median_filtered_dsg.median()

# Display the median values
print("Median Values for the Selected Columns:")
print(median_values_spnd)

print("Median Values for the Selected Columns:")
print(median_filtered_dsg)
```

```python
#Plotting standard deviation

import matplotlib.pyplot as plt

# Lists for columns for standard deviation by spend and dosage
stddev_columns_spnd = ['Total Spending 2019',
                      'Total Spending 2020',
                      'Total Spending 2021']

stddev_columns_dsg = ['Total Dsg Units 2019',
                      'Total Dsg Units 2020',
                      'Total Dsg Units 2021']

# Extract the columns of interest from your DataFrame
stddev_spnd = medi_filtered_df[stddev_columns_spnd].apply(pd.to_numeric, errors='coerce')
stddev_dsg = medi_filtered_df[stddev_columns_dsg].apply(pd.to_numeric, errors='coerce')

# Calculate standard deviation for spending and dosage
stddev_spnd_values = stddev_spnd.std()
stddev_dsg_values = stddev_dsg.std()

# Create line plots for standard deviation
plt.figure(figsize=(12, 6))

# Line plot for standard deviation of spending
plt.plot(stddev_spnd_values.index, stddev_spnd_values.values, label='Standard Deviation of Spending', marker='o')

# Line plot for standard deviation of dosage
plt.plot(stddev_dsg_values.index, stddev_dsg_values.values, label='Standard Deviation of Dosage', marker='o')

plt.title('Standard Deviation for Spending and Dosage (2019-2021)')
plt.xlabel('Year')
plt.ylabel('Standard Deviation')
plt.legend()
plt.ticklabel_format(style='plain', axis='y')
plt.grid(True)
plt.tight_layout()

# Save the plot as an image (e.g., in PNG format)
plt.savefig('Resources/standarddeviation_values_plot.png')
plt.show()

# Print the standard deviations
print('Standard Deviation of Spending:')
print(stddev_spnd_values)
print('\nStandard Deviation of Dosage:')
print(stddev_dsg_values)
```

```python
#Plot the top manufacturers of insulin between 2019 and 2021

# Who produced and distributed the most insulin through medicare from 
# 2019-2021? Who are the top 5 insulin producers? 

# Filter out rows containing 'overall' in 'Mftr_Name' column
medi_filtered_manufacture_df = medi_filtered_df[~medi_filtered_df['Mftr_Name'].str.contains('overall', case=False)]

# Filter the DataFrame for the years 2019, 2020, and 2021
years_to_include = ['Total Dsg Units 2019',
                   'Total Dsg Units 2020',
                   'Total Dsg Units 2021']
mftr_filtered_df = medi_filtered_df[['Mftr_Name'] + years_to_include]

# Group the data by manufacturer and calculate the total insulin production for each manufacturer
medi_grouped_df = mftr_filtered_df.groupby('Mftr_Name')[years_to_include].sum()

# Calculate the total production for each manufacturer
medi_grouped_df['Total_Production'] = medi_grouped_df.sum(axis=1)

# Sort the manufacturers by total production in descending order to identify the top producers
insulin_top_mftr = medi_grouped_df.sort_values(by='Total_Production', ascending=False).head(10)

# Create a bar chart to visualize the top insulin producers
plt.figure(figsize=(12, 10))
insulin_top_mftr['Total_Production'].plot(kind='bar')
plt.title('Top Insulin Producers (2019-2021)')
plt.xlabel('Manufacturer')
plt.ylabel('Total Production')
plt.ticklabel_format(style='plain', axis='y')
plt.xticks(rotation=90)
plt.tight_layout()

# Save the plot as an image 
plt.savefig('Resources/top_insulin_mftrs.png')

# Add a legend to the chart
plt.legend()
plt.show()
```
