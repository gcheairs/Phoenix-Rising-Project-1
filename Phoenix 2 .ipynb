```python
# Define relevant columns for the years 2019, 2020, and 2021
medi_relevant_columns = ['Gnrc_Name', 'Mftr_Name', 'Avg_Spnd_Per_Clm_2019', 'Avg_Spnd_Per_Clm_2020', 'Avg_Spnd_Per_Clm_2021']

# Filter rows based on 'Gnrc_Name' and 'Mftr_Name' conditions
filtered_df = medicare_insulin[medicare_insulin['Gnrc_Name'].isin(medi_names_to_filter) & ~medicare_insulin['Mftr_Name'].str.contains('Overall')]

# Created a copy of the DataFrame to avoid Setting With Copy Warning
medi_result_df = filtered_df[medi_relevant_columns].copy()

# Add dollar signs to 'Avg_Spnd_Per_Clm' columns
for year in ['2019', '2020', '2021']:
    medi_result_df[f'Avg_Spnd_Per_Clm_{year}'] = '$' + medi_result_df[f'Avg_Spnd_Per_Clm_{year}'].astype(str)

# Combine 'Avg_Spnd_Per_Clm' columns for the years 2019, 2020, and 2021
medi_result_df['Combined_Avg_Spnd_Per_Clm'] = medi_result_df[['Avg_Spnd_Per_Clm_2019', 'Avg_Spnd_Per_Clm_2020', 'Avg_Spnd_Per_Clm_2021']].agg(','.join, axis=1)

# Drop rows where any 'Avg_Spnd_Per_Clm' column is null
medi_result_df = medi_result_df.dropna(subset=[f'Avg_Spnd_Per_Clm_{year}' for year in ['2019', '2020', '2021']])

# Add a column indicating the years
medi_result_df['Years'] = medi_result_df.apply(lambda row: ', '.join([year for year in ['2019', '2020', '2021'] if row[f'Avg_Spnd_Per_Clm_{year}'] != '$nan']), axis=1)

# Sort by 'Combined_Avg_Spnd_Per_Clm' and get the top 5 claims
top_5_claims_df = medi_result_df.sort_values(by='Combined_Avg_Spnd_Per_Clm', ascending=False).head(5)

# Display the top 5 result DataFrame with only relevant columns
#print(top_5_claims_df[['Gnrc_Name', 'Mftr_Name', 'Combined_Avg_Spnd_Per_Clm', 'Years']])
```

```python
# Define relevant columns for the years 2019, 2020, and 2021
medi_relevant_columns = ['Gnrc_Name', 'Mftr_Name', 'Tot_Spndng_2019', 'Tot_Spndng_2020', 'Tot_Spndng_2021']

# Filter rows based on 'Gnrc_Name' and 'Mftr_Name' 
filtered_df = medicare_insulin[medicare_insulin['Gnrc_Name'].isin(medi_names_to_filter) & ~medicare_insulin['Mftr_Name'].str.contains('Overall')]


# Created a copy of the DataFrame to avoid Setting With Copy Warning
medi_result_df = filtered_df[medi_relevant_columns].copy()

# Adding dollar signs to 'Tot_Spndng' columns
for year in ['2019', '2020', '2021']:
    medi_result_df[f'Tot_Spndng_{year}'] = '$' + medi_result_df[f'Tot_Spndng_{year}'].astype(str)

# Convert 'Tot_Spndng' columns to numeric
for year in ['2019', '2020', '2021']:
    medi_result_df[f'Tot_Spndng_{year}'] = pd.to_numeric(medi_result_df[f'Tot_Spndng_{year}'].str.replace('$', ''), errors='coerce')

# Drop rows where any 'Tot_Spndng' column is null
medi_result_df = medi_result_df.dropna(subset=[f'Tot_Spndng_{year}' for year in ['2019', '2020', '2021']])

# Display the top 10 for each year with commas and a dollar sign
for year in ['2019', '2020', '2021']:
    top_10_df = medi_result_df.nlargest(10, f'Tot_Spndng_{year}')[['Gnrc_Name', f'Tot_Spndng_{year}']]
    top_10_df[f'Tot_Spndng_{year}'] = top_10_df[f'Tot_Spndng_{year}'].apply(lambda x: '${:,.0f}'.format(x))
    grand_total = top_10_df[f'Tot_Spndng_{year}'].str.replace('[$,]', '', regex=True).astype(float).sum()
    print(f'Top 10 for {year}:')
    print(top_10_df)
    print('Grand Total for {}: ${:,.0f}'.format(year, grand_total))
    print('\n')
```

```python
# Created a copy of the DataFrame to avoid Setting With Copy Warning
medi_result_df = filtered_df[medi_relevant_columns].copy()

# Add dollar signs to 'Tot_Spndng' columns
for year in ['2019', '2020', '2021']:
    medi_result_df[f'Tot_Spndng_{year}'] = '$' + medi_result_df[f'Tot_Spndng_{year}'].astype(str)

# Convert 'Tot_Spndng' columns to numeric
for year in ['2019', '2020', '2021']:
    medi_result_df[f'Tot_Spndng_{year}'] = pd.to_numeric(medi_result_df[f'Tot_Spndng_{year}'].str.replace('$', ''), errors='coerce')

# Drop rows where any 'Tot_Spndng' column is null
medi_result_df = medi_result_df.dropna(subset=[f'Tot_Spndng_{year}' for year in ['2019', '2020', '2021']])

# Group by 'Gnrc_Name' and sum the total spending for each year
grouped_df = medi_result_df.groupby('Gnrc_Name')[[f'Tot_Spndng_{year}' for year in ['2019', '2020', '2021']]].sum()

# Define different colors for each bar
colors = ['blue', 'purple', 'gold']

# Plot a bar graph for each year with distinct colors
for year, color in zip(['2019', '2020', '2021'], colors):
    plt.figure(figsize=(10, 6))
    grouped_df[f'Tot_Spndng_{year}'].sort_values(ascending=False).head(10).plot(kind='bar', color=color)
    plt.title(f'Top 10 Total Spending for {year}')
    plt.xlabel('Generic Name')
    plt.ylabel('Total Spending ($)')
    plt.xticks(rotation=45, ha='right')
    plt.show()
```

```python
# Define relevant columns for the years 2019, 2020, and 2021
medi_relevant_columns = ['Gnrc_Name', 'Mftr_Name', 'Tot_Benes_2019', 'Tot_Benes_2020', 'Tot_Benes_2021']

# Filter rows based on 'Gnrc_Name' and 'Mftr_Name'
filtered_df = medicare_insulin[medicare_insulin['Gnrc_Name'].isin(medi_names_to_filter) & ~medicare_insulin['Mftr_Name'].str.contains('Overall')]

# Created a copy of the DataFrame to avoid Setting With Copy Warning
medi_result_df = filtered_df[medi_relevant_columns].copy()

# Convert 'Tot_Benes' columns to numeric without removing dollar signs
for year in ['2019', '2020', '2021']:
    medi_result_df[f'Tot_Benes_{year}'] = pd.to_numeric(medi_result_df[f'Tot_Benes_{year}'], errors='coerce')

# Drop rows where any 'Tot_Benes' column is null
medi_result_df = medi_result_df.dropna(subset=[f'Tot_Benes_{year}' for year in ['2019', '2020', '2021']])

# Display the top 10 for each year without decimal places and with commas
for year in ['2019', '2020', '2021']:
    top_10_df = medi_result_df.nlargest(10, f'Tot_Benes_{year}')[['Gnrc_Name', f'Tot_Benes_{year}']]
    top_10_df[f'Tot_Benes_{year}'] = top_10_df[f'Tot_Benes_{year}'].apply(lambda x: '{:,.0f}'.format(x))
    grand_total = top_10_df[f'Tot_Benes_{year}'].str.replace(',', '').astype(float).sum()
    print(f'Top 10 for {year}:')
    print(top_10_df)
    print('Grand Total for {}: {:,.0f}'.format(year, grand_total))
    print('\n')
```

```python
# Create a copy of the DataFrame to avoid Setting With Copy Warning
medi_result_df = filtered_df[medi_relevant_columns].copy()

# Add dollar signs to 'Tot_Benes' columns
for year in ['2019', '2020', '2021']:
    medi_result_df[f'Tot_Benes_{year}'] = '$' + medi_result_df[f'Tot_Benes_{year}'].astype(str)

# Convert 'Tot_Benes' columns to numeric
for year in ['2019', '2020', '2021']:
    medi_result_df[f'Tot_Benes_{year}'] = pd.to_numeric(medi_result_df[f'Tot_Benes_{year}'].str.replace('$', ''), errors='coerce')

# Drop rows where any 'Tot_Benes' column is null
medi_result_df = medi_result_df.dropna(subset=[f'Tot_Benes_{year}' for year in ['2019', '2020', '2021']])

# Group by 'Gnrc_Name' and sum the total beneficiaries for each year
grouped_df = medi_result_df.groupby('Gnrc_Name')[[f'Tot_Benes_{year}' for year in ['2019', '2020', '2021']]].sum()

# Print the columns to check their names
print(grouped_df.columns)


# Define different colors for each bar
colors = ['red', 'green', 'brown']

# Plot a bar graph for each year
for year, color in zip(['2019', '2020', '2021'], colors):
    plt.figure(figsize=(8, 6))
    grouped_df.sort_values(by=f'Tot_Benes_{year}', ascending=False).head(10)[f'Tot_Benes_{year}'].plot(kind='bar', color=color)
    plt.title(f'Top 10 Total Beneficiaries for {year}')
    plt.xlabel('Generic Name')
    plt.ylabel('Total Beneficiaries')
    plt.xticks(rotation=45, ha='right')
    plt.show()
```

```python
# Define relevant columns for the years 2019, 2020, and 2021
medi_relevant_columns = ['Gnrc_Name', 'Mftr_Name', 'Tot_Benes_2019', 'Tot_Benes_2020', 'Tot_Benes_2021', 'Tot_Spndng_2019', 'Tot_Spndng_2020', 'Tot_Spndng_2021']

# Filter rows based on 'Gnrc_Name' and 'Mftr_Name'
filtered_df = medicare_insulin[medicare_insulin['Gnrc_Name'].isin(medi_names_to_filter) & ~medicare_insulin['Mftr_Name'].str.contains('Overall')]

# Create a copy of the DataFrame to avoid Setting With Copy Warning
medi_result_df = filtered_df[medi_relevant_columns].copy()

# Convert 'Tot_Benes' and 'Tot_Spndng' columns to numeric without removing dollar signs
for col in ['Tot_Benes_2019', 'Tot_Benes_2020', 'Tot_Benes_2021', 'Tot_Spndng_2019', 'Tot_Spndng_2020', 'Tot_Spndng_2021']:
    medi_result_df[col] = pd.to_numeric(medi_result_df[col].astype(str).str.replace('$', '').str.replace(',', ''), errors='coerce')

# Drop rows where any relevant column is null
medi_result_df = medi_result_df.dropna(subset=['Tot_Benes_2019', 'Tot_Benes_2020', 'Tot_Benes_2021', 'Tot_Spndng_2019', 'Tot_Spndng_2020', 'Tot_Spndng_2021'])

# Calculate the ratio of 'Tot_Spndng' to 'Tot_Benes' for each year
for year in ['2019', '2020', '2021']:
    medi_result_df[f'Ratio_{year}'] = medi_result_df[f'Tot_Spndng_{year}'] / medi_result_df[f'Tot_Benes_{year}']

# Display the top 10 for each year with commas and a dollar sign
for year in ['2019', '2020', '2021']:
    top_10_df = medi_result_df.nlargest(10, f'Ratio_{year}')[['Gnrc_Name', f'Ratio_{year}']]
    top_10_df[f'Ratio_{year}'] = top_10_df[f'Ratio_{year}'].apply(lambda x: '${:,.2f}'.format(x))
    grand_total = medi_result_df[f'Tot_Spndng_{year}'].sum() / medi_result_df[f'Tot_Benes_{year}'].sum()
    print(f'Top 10 for {year}:')
    print(top_10_df)
    print(f'Grand Total for {year}: ${grand_total:.2f}\n')
```
